```python
import pandas as pd

# Load datasets with correct mapping
customers = pd.read_csv("data/olist_customers_dataset.csv")
geolocation = pd.read_csv("data/olist_geolocation_dataset.csv")
order_items = pd.read_csv("data/olist_order_items_dataset.csv")
order_payments = pd.read_csv("data/olist_order_payments_dataset.csv")
order_reviews = pd.read_csv("data/olist_order_reviews_dataset.csv")
orders = pd.read_csv("data/olist_orders_dataset.csv")
products = pd.read_csv("data/olist_products_dataset.csv")
sellers = pd.read_csv("data/olist_sellers_dataset.csv")
category_translation = pd.read_csv("data/product_category_name_translation.csv")

# Step-by-step merging
df = orders.merge(customers, on='customer_id', how='left')
df = df.merge(order_items, on='order_id', how='left')
df = df.merge(products, on='product_id', how='left')
df = df.merge(category_translation, on='product_category_name', how='left')
df = df.merge(sellers, on='seller_id', how='left')
df = df.merge(order_payments, on='order_id', how='left')
df = df.merge(order_reviews, on='order_id', how='left')

# Total price
df['total_price'] = df['price'] + df['freight_value']

# Convert all relevant date columns
date_columns = [
    'order_purchase_timestamp', 'order_approved_at', 'order_delivered_carrier_date',
    'order_delivered_customer_date', 'order_estimated_delivery_date',
    'shipping_limit_date', 'review_creation_date', 'review_answer_timestamp'
]
for col in date_columns:
    if col in df.columns:
        df[col] = pd.to_datetime(df[col])

# Preview
print("‚úÖ Master dataset created successfully!")
print("Shape:", df.shape)
df.head()
```

```python
# Remove duplicates to count unique entities at order/customer/seller level
unique_orders = df['order_id'].nunique()
unique_customers = df['customer_id'].nunique()
unique_sellers = df['seller_id'].nunique()
unique_products = df['product_id'].nunique()
total_reviews = df['review_id'].nunique()

# Revenue (sum of product price + freight)
total_revenue = df['total_price'].sum()

# Average order value (AOV)
aov = df.groupby('order_id')['total_price'].sum().mean()

# Average items per order
avg_items_per_order = df.groupby('order_id')['order_item_id'].count().mean()

# Unique payment methods
unique_payment_types = df['payment_type'].nunique()

# Compose the metrics in a summary DataFrame
kpi_summary = pd.DataFrame({
    'Metric': [
        'Total Orders',
        'Total Customers',
        'Total Sellers',
        'Total Unique Products',
        'Total Revenue (BRL)',
        'Average Order Value (BRL)',
        'Average Items per Order',
        'Total Reviews',
        'Unique Payment Methods'
    ],
    'Value': [
        unique_orders,
        unique_customers,
        unique_sellers,
        unique_products,
        round(total_revenue, 2),
        round(aov, 2),
        round(avg_items_per_order, 2),
        total_reviews,
        unique_payment_types
    ]
})

# Show the summary
import pandas as pd
import matplotlib.pyplot as plt

display(kpi_summary)
```

```python
import pandas as pd

# Load your dataset
df = pd.read_csv("master_ecommerce_dataset.csv")

df['review_score'].fillna(df['review_score'].mean(), inplace=True)


# Impute average review_score in missing cells
avg_score = df["review_score"].mean()
df["review_score"] = df["review_score"].fillna(round(avg_score, 1))

# Save the updated file
df.to_csv("master_ecommerce_dataset_cleaned.csv", index=False)
print("‚úÖ Missing review_score values filled with average and file saved.")
```
